```python
medals_new = medals.drop(['City'],axis=1)
```

```python
Eques_medals = medals_new.loc[(medals_new.Sport=='Equestrian')&(medals_new.Event=='Team'),:]
Eques_medals.head(40)
```

```python
Eques_pivot = Eques_medals.pivot_table(index=['Year','Country','Sport','Discipline'],
                       columns='Medal',
                       values='Athlete',
                       aggfunc='count',
                       fill_value=0)
Eques_pivot
```
